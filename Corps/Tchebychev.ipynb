## Polynômes de tchebychev de premier degré
Elles sont définies par la relation de récurrence : 

$T_0(X)=1$

$T_1(X)=X$

$T_{n+2}(X)=2X T_{n+1}(X) - T_n(X)$

par ailleurs : 

$T_2(X) = 2 X^2 - 1$

$T_3(X) = 4 X^3 - 3 X$

$T_4(X) = 8 X^4 - 8 X^2 + 1$

```python
T_0 = 1* X^0
T_1 = X
T_2 = 2*X^2 - 1
T_3 = 4*X^3 - 3*X
T_4 = 8*X^4 - 8*X^2 + 1
```

## Polynômes de Tchebychev de seconde espèce

Elles sont définies par : 

$U_0 = $

$U_1 = 2X$

$U_{n+2}= 2 X U_{n+1} - U_n $

Par ailleurs : 


$U_2 = 4 X^2 -1 $

$U_3 = 8 X^3 - 4X$

$U_4 = 16 X^4 - 12 X^2 +1$

```python
#On les programme directement récursivement : 

U_0 = T_0
U_1 = 2*X

def Tchebychev2(n):
    if n==0:
        return U_0
    if n==1:
        return U_1
    else : 
        return 2*X*Tchebychev2(n-1)- Tchebychev2(n-2)
```
