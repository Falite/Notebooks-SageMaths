# Equations de Bring Jerard

Il s'agit des équations du type 

$x^5+x+a=0$   avec $a$ une constante rationelle.

Pour s'y faire, l'on va écrire ces équations de la forme $P(x)=0$ avec $P \in \mathbb Q[X]$

```python
%display latex
```

#### Avant de commencer
On va se servir de plusieurs outils qui nous aideront pour les affichages graphiques

```python
import networkx as nx
import hashlib
```

```python
#génarateur de couleurs 
def generateurcouleurs(i):
    hex_dig=hashlib.md5(str(i).encode()).hexdigest()
    return f"#{hex_dig[:6]}"
```

### Construction des objets qui vont nous servir

```python
PRQ.<X> = PolynomialRing(QQ)
PRR.<Y> = PolynomialRing(RR)
var('x') #pour les affichages
```

```python
#On définit une fonction qui nous renverra un polynôme de Bring-Jerard
def BringJerard(a):
    P = X^5+X+a
    return P

def BringJerardReel(a):
    P = Y^5+Y+a
    return P
```

## Exemple d'un polynôme de Bring-Jerard
#### Travaux sur un polynôme de Bring-Jerard

On va commencer par le cas $a=5$, c'est à dire le polynôme $P=X^5+X+5$

```python
P = BringJerard(5)
P.factor()
```

#### Affichage

```python
tr_P = x^5+x+5
plot(tr_P,(x,-5,5),
     ymin=0,ymax=30,color="blue")
```

```python
P.is_irreducible()
```

On a $P$ est bien un polynôme irréductible.

#### Recherche de ses racines réelles :

```python
Pr = BringJerardReel(5)
Pr.roots()
```

Ce polynôme admet donc une seule racine réelle et deux paires de racines complexes conjuguées.

#### Application de la théorie de Galois sur ce polynôme :

On veut d'abord calculer le corps $\frac{\mathbb Q [X] }{(P)_\text{id}}$.


On sait que ça sera un corps car $P$ est irréductible.

```python
K.<alpha> = NumberField(P)
print(K)
```

Calculons maintenant son groupe de Galois à partir de son corps de décomposition :

```python
G = K.galois_group()
print(G)
```

Ce groupe est en bijection avec $S_5$ le groupe des permutations entre 5 éléments. On sait qu'il n'est pas résoluble.

## Travaux sur plusieurs polynômes de Bring Jerard

On va commencer par afficher plusieurs polynômes en faisant varier $a$ de 1 à 20.

```python
tracés=[]
for i in range(1,21):
    couleur=generateurcouleurs(i)
    tracé = plot(x^5+x+i,(x,-5,5),
                 ymin=0,ymax=30,color=couleur)
    tracés.append(tracé)

tracé=sum(tracés)
tracé.show(figsize=(10,6))
```

On peut voir qu'elles semblent admettre une seule racine réelles.

#### Recherche des racines réelles et irréductibilité des polynômes

```python
Polynômes=[]
for i in range(1,21):
    Polynômes.append(BringJerard(i))

for P in Polynômes:
    if P.is_irreducible():
        print(P," est irréductible")
    else:
        print(P," n'est pas irréductible")
```

Elles ne sont pas toutes irréductibles. 

```python
for i in range(1,21):
    print("les racines réelles de ",BringJerard(i),"sont :",BringJerardReel(i).roots())
```

Comme on l'avait dit, elles admettent une seule racine réelle de multiplicité 1.

### Application de la théorie de Galois sur ces polynômes :
Nous allons d'abord calculer $\frac{\mathbb Q[X]}{(P)_{id}}$ uniquement si $P$ est irréductible.

Pour s'y faire, la méthode `.is_irreducible`va nous être très utile.

```python
for P in Polynômes:
    if P.is_irreducible():
        K.<alpha> = NumberField(P)
        print(K)
    else:
        print(P,"n'est pas irréductible")
```

Calculons maintenant les groupes de Galois et voyons s'ils sont récolvables ou non

```python
for P in Polynômes:
    if P.is_irreducible():
        K.<alpha> = NumberField(P)
        G = K.galois_group()
        print(G)
```

Ils sont tous en bijection avec $S_5$ le groupe des permutations entre 5 éléments. On sait qu'il n'est pas résoluble.
