## Premiers Test de Sage pour les extensions de corps 
L'on va tout simplement coder l'exemple 16 du cours :
L'extension $\mathbb Q ( 2^{\frac 13 } )$ :                
En effet dans $\mathbb R$, l'Analyse montre que $\exists ! a \in \mathbb R , ~a^3 = 2~~$.            
Et ce nombre $a$ est algébrique sur $\mathbb Q$ puisque racine du polynôme $X^3-2 ~\in \mathbb Q [X]$

```python
%display latex
import networkx as nx
import matplotlib.pyplot as plt
```

```python
K = QQ
```

```python
X = polygen(K)
P = X^3 - 2
```

```python
P.roots()
```

On a défini P sur K=$\mathbb Q[X]$ , par la méthode `.roots()`, on voit qu'elle n'a pas de racines dans $K$

```python
P.factor()
```

De plus, par la méthode `.factor()` , on remarque que P ne peut être plus factorisé 

```python
G = P.galois_group()
G
```

On peut construire le groupe de Galois avec la méthode `.galois_group()`

```python
print("groupe de galois :" , G ) 
```

```python
print(G.subgroups())
```

```python
elements = G.list()
elements
```

```python
graph = nx.Graph()
for elem in elements:
    graph.add_node(elem)
for i in range(len(elements)):
    for j in range(i+1,len(elements)):
        graph.add_edge(elements[i],elements[j])
pos = nx.spring_layout(graph)
nx.draw(graph, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_color='black')
plt.show()
```
